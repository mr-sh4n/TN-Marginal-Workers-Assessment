```python
#data preprocessing
```

```python
#check for missing values
print(df.isnull().sum())
```

```python
#handle missing values
df=df.dropna()
```

```python
#remove duplicates
df=df.drop_duplicates()
```

```python
#convert data types
#df['column_name']=df['column_name'].astype('desired_type')
```

```python
#save the dataframe to a new CSV file
df.to_csv(r'C:\Users\fathi\OneDrive\Documents\cleaned_dataset.csv',index=False)
```

```python
#demographic analysis
#Calculate the distribution of marginal workers based on age, industrial category, and sex
```

```python
#group by age, industrial category, and sex, and calculate the counts
age_distribution=df.groupby('Age group').size()
industrial_category_distribution=df.groupby('Industrial Category - A - Cultivators - Persons').size()
area_distribution=df.groupby('Area Name').size()
```

```python
#Data Visualization
```

```python
#Age Distribution
plt.figure(figsize=(10,6))
sns.barplot(x=age_distribution.index,y=age_distribution.values)
plt.title('Age Distribution of Marginal Workers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#industriaal Category distribution
plt.figure(figsize=(10,6))
sns.barplot(x=industrial_category_distribution.index,y=industrial_category_distribution.values)
plt.title('Industrial Category Distribution of Marginal Workers')
plt.xlabel('Industrial Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#Area Distribution
plt.figure(figsize=(6,6))
sns.barplot(x=area_distribution.index,y=area_distribution.values)
plt.title('Area Distribution of Marginal Workers')
plt.xlabel('Area')
plt.ylabel('Count')
plt.show()
```
